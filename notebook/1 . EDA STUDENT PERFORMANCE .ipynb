### 3.8 Adding columns for "Total Score" and "Average"

```python
df['total score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total score']/3
df.head()
```

```python
reading_full = (df['reading_score'] == 100).sum()
writing_full = (df['writing_score'] == 100).sum()
math_full = (df['math_score'] == 100).sum()

print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
```

```python
reading_less_20 = (df['reading_score'] <= 20).sum()
writing_less_20 = (df['writing_score'] <= 20).sum()
math_less_20 = (df['math_score'] <= 20).sum()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```

#####  Insights
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section

#### 4.3 Multivariate analysis using pieplot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(30, 15))

# --- Gender ---
plt.subplot(2, 3, 1)
sizes = df['gender'].value_counts()
plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%')
plt.title('Gender', fontsize=20, fontweight="bold")

# --- Race/Ethnicity ---
plt.subplot(2, 3, 2)
sizes = df['race_ethnicity'].value_counts()
plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%')
plt.title('Race / Ethnicity', fontsize=20, fontweight="bold")

# --- Lunch ---
plt.subplot(2, 3, 3)
sizes = df['lunch'].value_counts()
plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%')
plt.title('Lunch', fontsize=20, fontweight="bold")

# --- Test Preparation ---
plt.subplot(2, 3, 4)
sizes = df['test_preparation_course'].value_counts()
plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%')
plt.title('Test Preparation', fontsize=20, fontweight="bold")

# --- Parental Education ---
plt.subplot(2, 3, 5)
sizes = df['parental_level_of_education'].value_counts()
plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%')
plt.title("Parental Education", fontsize=20, fontweight="bold")

plt.tight_layout()
plt.show()
```

#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? ) 

```python
gender_group = df.groupby('gender').mean(numeric_only=True)
print(gender_group)
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average', 'Math Average']

female_scores = [gender_group.loc['female', 'average'], gender_group.loc['female', 'math_score']]
male_scores   = [gender_group.loc['male', 'average'], gender_group.loc['male', 'math_score']]

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, male_scores, 0.4, label='Male', color='skyblue')
plt.bar(X_axis + 0.2, female_scores, 0.4, label='Female', color='lightcoral')

plt.xticks(X_axis, X, fontsize=12, fontweight="bold")
plt.ylabel("Marks", fontsize=12, fontweight="bold")
plt.title("Total Average vs Math Average Marks of Both Genders", fontsize=14, fontweight="bold")
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.
